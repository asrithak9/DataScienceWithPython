```python
import random
import string
import matplotlib.pyplot as pl

#Functoion to generate random strings  

def randomGenerator(length) :
    characters = string.printable
    return '' .join(random.choice(characters) for x in range(length))


randomList = []
def randomStringGenerator(numberofStrings) :
    i=1
    for i in range(numberofStrings):
        j = random.randint(1, 20)
        randomList.append(randomGenerator(j))

#Function to generate ascii value of string

def asciGenerator(string_name) :
    asciofString = 0
    for element in range(0, len(string_name)): 
        c = ord(string_name[element])
        asciofString = asciofString + c
        
    return asciofString


#Sorting function

def bubble_sort():
    l = len(randomList)
    print('Strings before sorting', randomList)

    # Traverse through array elements
    for i in range(l):
 
         for j in range(0, l-i-1):
 
            # traverse the array from 0 to l-i-1
            # Swap if the strings's value is greater
            # than the next element
            if asciGenerator(randomList[j]) > asciGenerator(randomList[j+1]) :
                randomList[j], randomList[j+1] = randomList[j+1], randomList[j]  
     
    print('\nStrings after sorting', randomList)
    randomList.clear()
    

# Test case 1 : Sorting 2 strings
print("********TEST CASE 1**********")
print("Sorting 2 strings")
randomStringGenerator(2)
bubble_sort()


# Test case 2 : Sorting 5 strings
print("\n\n********TEST CASE 2**********")
print("Sorting 5 strings")
randomStringGenerator(5)
bubble_sort()


# Test case 3 : Sorting 7 strings
print("\n\n********TEST CASE 3**********")
print("Sorting 7 strings")
randomStringGenerator(7)
bubble_sort()

# Test case 4 : Sorting 10 strings
print("\n\n********TEST CASE 4**********")
print("Sorting 10 strings")
randomStringGenerator(10)
bubble_sort()

# Test case 5 : Sorting 9 strings
print("\n\n********TEST CASE 5**********")
print("Sorting 9 strings")
randomStringGenerator(9)
bubble_sort()
```

```python
#Scatter plot with length of string on x-axis and order of string in list on y-axis

randomStringGenerator(2)
l_xaxis = []
o_yaxis = []
for s in randomList :
    l_xaxis.append(len(str(s)))
    o_yaxis.append(randomList.index(s))

pl.figure(figsize=(20,10))
pl.scatter(o_yaxis, l_xaxis)
pl.show()
```
