```
import random
import numpy as np
import matplotlib.pyplot as plt

X_Coordinates = []
Y_Coordinates = []
x_in_circle = []
y_in_circle = []
x_out_circle = []
y_out_circle = []
pi_values = []
radius = 0.5

#To determine the position of a point
def PointsInCircle(no_of_points):
  points_in_circle = 0
  i=0
  X_Coordinates = np.array([random.uniform(0,1) for _ in range(no_of_points)])
  Y_Coordinates = np.array([random.uniform(0,1) for _ in range(no_of_points)])
  for i in range(no_of_points):
    distance = (X_Coordinates[i]-0.5)**2 + (Y_Coordinates[i]-0.5)**2
    if distance <= radius*radius:
        points_in_circle += 1
        x_in_circle.append(X_Coordinates[i])
        y_in_circle.append(Y_Coordinates[i])
    else:
        x_out_circle.append(X_Coordinates[i])
        y_out_circle.append(Y_Coordinates[i])
  return points_in_circle,x_in_circle,y_in_circle,x_out_circle,y_out_circle
```

```
# Graph plot for different points in simulation
def plot(Xin,Yin,Xout,Yout):
  circle = plt.Circle((0.5, 0.5), radius, color='black',fill=False)
  fig, ax = plt.subplots(figsize=(7,7))
  plt.xlim(0,1)
  plt.ylim(0,1)
  ax.add_artist(circle)
  plt.scatter(x_in_circle,y_in_circle,color="red")
  plt.scatter(x_out_circle,y_out_circle,color="blue")
  plt.show()
```

```
#Monte Carlo method
pi_values = []
j=0
list_of_no_of_points = [10,500,1000,5000,10000,50000,80000,100000,200000]
for j in range(len(list_of_no_of_points)):
  points,xin,yin,xout,yout = PointsInCircle(list_of_no_of_points[j])
  pivalue = 4 * points / list_of_no_of_points[j]
  plot(xin,yin,xout,yout)
  pi_values.append(pivalue)

print('\n\nPi values in given list:',pi_values)
```
